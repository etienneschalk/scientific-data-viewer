```python
import xarray as xr

xds = xr.open_dataset(
    "https://data.earthdatahub.destine.eu/era5/era5-pressure-levels-for-wind-developers-v0.zarr",
    storage_options={"client_kwargs":{"trust_env":True}},
    chunks={},
    engine="zarr",
)
xds
```

```python
xds["r"].isel(isobaricInhPa=0).isel(valid_time=0).plot()
```
